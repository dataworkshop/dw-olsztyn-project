```
tupla = input("Podaj liczby oddzielone przecinkami: ")
NuberrArr = tupla.split(',')
lista = [] # deklaracja pustej listy
for i in range(len(NuberrArr)):
    lista.append(int(NuberrArr[i]))

print ("Elementy i ich indeksy:")
for i, v in enumerate(lista):
    print ("element: "+ str(v) +"  jego indeks:" + str(i))

print ("Elementy w odwrÃ³conym porzÄ…dku:")
for e in reversed(lista):
    print (e),

print ("")
print ("Elementy posortowane rosnÄ…co:")
for e in sorted(lista):
    print (e),

print ("")
e = int(input("KtÃ³rÄ… liczbÄ™ usunÄ…Ä‡? "))
lista.remove(e)
print (lista)

a  = int(input("Podaj element do dodania: "))
i = int (input("Podaj indeks, przed ktÃ³rym ma sie on znaleÅºÄ‡: "))
lista.insert(i, a)
print (lista)

e = int(input("Podaj liczbÄ™, ktÃ³rej wystÄ…pienia w liÅ›cie chcesz zliczyÄ‡? "))
print (lista.count(e))
print ("Pierwszy indeks, pod ktÃ³rym zapisana jest podana liczba to: ")
print (lista.index(e))

print ("Pobieramy ostatni element z listy: ")
print (lista.pop())
print (lista)

i = int(input("Podaj indeks poczÄ…tkowy:"))
j =int(input("Podaj indeks koÅ„cowy, aby uzyskaÄ‡ frgament listy: "))
print (lista[i:j])
```

JeÅ¼eli chcemy trwale odwrÃ³ciÄ‡ lub posortowaÄ‡ elementy listy stosujemy metody: .reverse() i .sort().
